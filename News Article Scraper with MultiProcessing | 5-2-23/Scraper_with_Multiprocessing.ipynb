```python
import os
import time
import requests
import numpy as np
from tqdm import tqdm
from bs4 import BeautifulSoup
from multiprocessing import Process

categories = {  "Politics"  : 'https://theprint.in/category/politics/', # 1727
                "Economy"   : 'https://theprint.in/category/economy/' , # 1483
                "Defense"   : 'https://theprint.in/category/defence/',  # 287
                "India"     : 'https://theprint.in/category/india/',    # 17319
                "Health"    : 'https://theprint.in/category/health/',   # 793
                "Science"   : 'https://theprint.in/category/science/',  # 335
                "Tech"      : 'https://theprint.in/category/tech/'}     # 445  
```

### 1. Finding the number of Pages

```python
links = []
pages = []

for key in categories.keys():

  link = categories[key]
  r = requests.get(link)
  soup = BeautifulSoup(r.text, 'html.parser')

  page = int(soup.find('a', class_ = 'last').text.replace(',',''))
  
  pages.append(page)
  links.append(link)
```

### 2. Defining Link Scraper

```python
def scrape_links(link, pages, category):

  links = []
  # pages = 100

  for page in tqdm(range(1, pages + 1)):
    lnk = link + 'page/' + str(page)
    r = requests.get(lnk)
    soup = BeautifulSoup(r.text,'html.parser')
    for sp in soup.find_all('div',class_ = 'tdb_module_loop td_module_wrap td-animation-stack '):
      links.append(sp.find('a').get('href'))

    time.sleep(.5)

  links = np.array(links)
  np.save(category + '.npy',links)
  print(category + '.npy file is saved!')
```

### 3. Calling Link Scraper

```python
for link, page in zip(links, pages):

  category = link.split('/')[-2]
  scrape_links(link, page, category)
```

### 4. Scraper with MultiProcessing

```python
processes = []

for link , page in zip(links, pages):
  category = link.split('/')[-2]
  processes.append(Process(target = scrape_links , args=(link,page,category)))

for process in processes:
  process.start()

for process in processes:
  process.join()
```
